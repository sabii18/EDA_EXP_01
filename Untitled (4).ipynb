```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].dt.year
```

```python
def win_type(row):

    if row['result'] == 'tie':
        return "Tie"

    elif row['result'] == 'wickets':
        return "Won by Wickets"

    elif row['result'] == 'runs':
        return "Won by Runs"

    else:
        return "No Result"

df['win_type'] = df.apply(win_type,axis=1)
```
